### âœ… Step 2: Mount the MCP App

```python
from fastapi import FastAPI
from mcp.server import create_app

app = FastAPI()
mcp_app = create_app()

# Mount MCP under "/mcp"
app.mount("/mcp", mcp_app)
```

Now, your MCP endpoints are available at:
