# 💾 **Installation** – *Set Up the MCP Python SDK*

---

## 📌 What It Does

The **Installation step** sets up the MCP SDK so you can start building tools, prompts, and run your own MCP server — either locally or inside a larger ASGI app.

---

## 🧪 Quick Installation (Recommended)

Just run:

```bash
pip install mcp
```

This installs:

* The **MCP SDK**
* All required dependencies for tools, prompts, context, completions, etc.

---

## 🛠️ Optional: Full Installation with Dev Tools

If you want all CLI tools, dev utilities, and integrations:

```bash
pip install "mcp[dev]"
```

This gives you:

* 🧪 Dev server support
* 🔬 CLI tools like `mcp run`, `mcp serve`
* 🧵 Threaded tool chaining
* 🖥️ Claude Desktop compatibility

---

## 📁 Installing from GitHub (Latest Version)

If you want bleeding-edge updates:

```bash
pip install git+https://github.com/modelcontextprotocol/python-sdk.git
```

---

## 💡 Requirements

| Requirement  | Version                 |
| ------------ | ----------------------- |
| Python       | 3.8 or newer            |
| OS Support   | macOS, Linux, Windows   |
| Dependencies | Auto-installed with pip |

---

## ⚠️ Troubleshooting Tip

If you hit issues with build tools or native extensions, make sure to update `pip`, `setuptools`, and `wheel`:

```bash
pip install --upgrade pip setuptools wheel
```

---
