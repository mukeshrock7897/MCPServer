## üìå What It Does

The MCP Python SDK allows you to **start and serve your own AI agent server**, locally or in production. You can run the server in development mode or mount it to existing web apps.

| Scenario                   | Why Use It                                    |
| -------------------------- | --------------------------------------------- |
| üß™ Local testing           | Rapidly prototype and test agents/tools       |
| üåê Cloud deployment        | Host MCP server behind web APIs or gateways   |
| üß© Integrate into web apps | Serve agents alongside your Flask/FastAPI app |

## ‚öôÔ∏è Key Methods to Run the Server

### ‚úÖ 1. Quick Start (Local Dev)

```bash
mcp server serve
```

This command **spins up your local MCP server** on the default port `3000`.

---

### ‚úÖ 2. Python Programmatic Start

```python
from mcp.server import MCPServer

server = MCPServer()
server.serve()
```

---

### ‚úÖ 3. Mount Inside an ASGI App (e.g., FastAPI, Starlette)

```python
from mcp.server import MCPServer

mcp_server = MCPServer()
app.mount("/mcp", mcp_server.app)
```

---

### ‚úÖ 4. Uvicorn / Gunicorn for Prod Hosting

```bash
uvicorn your_module:app --host 0.0.0.0 --port 8000
```

Use this when deploying in cloud platforms like **AWS**, **Render**, **GCP**, or **VPS**.
